```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r'C:\Users\user\Downloads\archive\Walmart_sales.csv'
df = pd.read_csv(file_path)

# Check missing values
print("Missing values per column before handling:")
print(df.isnull().sum())

# Fill missing values
df.fillna(df.select_dtypes(include=['number']).mean(), inplace=True)

# If non-numeric columns have missing values, use the mode
for column in df.select_dtypes(exclude=['number']).columns:
    df[column].fillna(df[column].mode()[0], inplace=True)

# Verify missing values are handled
print("\nMissing values per column after handling:")
print(df.isnull().sum())
```

```python
import pandas as pd

# Load the dataset
file_path = r'C:\Users\user\Downloads\archive\Walmart_sales.csv'
try:
    df = pd.read_csv(file_path)
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
    exit()

# Display basic statistics for numerical columns
print("\nBasic Statistics of Numerical Columns:")
print(df.describe())
```

```python
# Group by 'Holiday_Flag' and calculate multiple metrics for 'Weekly_Sales'
holiday_stats = df.groupby('Holiday_Flag')['Weekly_Sales'].agg(['mean', 'median', 'std'])

print("\nStatistics for Weekly Sales (Grouped by Holidays):")
print(holiday_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'Date' with error handling for mixed formats
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# Check rows where conversion failed
invalid_dates = df[df['Date'].isnull()]
print("Rows with invalid date format:")
print(invalid_dates)
df = df.dropna(subset=['Date'])

# Convert 'Date' to datetime format if not already done
df['Date'] = pd.to_datetime(df['Date'])

# Aggregate weekly sales by date
sales_trends = df.groupby('Date')['Weekly_Sales'].sum()

# Plot the line chart
plt.figure(figsize=(12, 6))
plt.plot(sales_trends.index, sales_trends.values, marker='o', linestyle='-', color='blue')
plt.title('Weekly Sales Trends Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Weekly Sales', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Group by 'Holiday_Flag' and calculate average sales
holiday_sales = df.groupby('Holiday_Flag')['Weekly_Sales'].mean()

# Plot the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=holiday_sales.index, y=holiday_sales.values, palette='viridis')
plt.title('Average Weekly Sales: Holidays vs Non-Holidays', fontsize=16)
plt.xlabel('Holiday (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Average Weekly Sales', fontsize=12)
plt.xticks([0, 1], ['Non-Holidays', 'Holidays'])
plt.show()
```

```python
# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Weekly_Sales'], kde=True, color='green', bins=30)
plt.title('Distribution of Weekly Sales', fontsize=16)
plt.xlabel('Weekly Sales', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# Plot the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature', y='Weekly_Sales', data=df, hue='Holiday_Flag', palette='coolwarm', alpha=0.7)
plt.title('Temperature vs Weekly Sales', fontsize=16)
plt.xlabel('Temperature (F)', fontsize=12)
plt.ylabel('Weekly Sales', fontsize=12)
plt.legend(title='Holiday')
plt.show()
```
